# Hi, This is the first code. Here we are trying to impute the missing UserIDs

```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('VisitorLogsData.csv')
```

```python
df['Activity'] = df['Activity'].str.upper()
df['ProductID'] = df['ProductID'].str.upper()
df['OS']= df['OS'].str.upper()
df['Browser']= df['Browser'].str.upper()
df['City']= df['City'].str.upper()
df['Country']= df['Country'].str.upper()
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
df.isnull().sum()
```

### Here we are trying to fill the null values in UserID based on the webClientID.

```python
web_user = df[['webClientID','UserID']]
web_user.dropna(inplace=True, axis =0)
web_user.drop_duplicates(inplace=True)
```

```python
df.loc[df.UserID.isnull(),'UserID'] = web_user.groupby(['webClientID'] )['UserID'].transform(lambda x: x)
```

```python
df.isnull().sum()
```

##### Turns out that the null elements are still the same. The webClientID corresponding to null users are all non-registered users. We are going to drop the rows.

### Dropping the null values in UserID and saving it to a new csv_file for code-2

```python
df = df.loc[df['UserID'].notnull()]
```
