### Creating a function to convert UNIX datetime to regular format of DateTime

##### The unix datetime seems to be in nano-seconds. So, we divide the unix value with 1e9

```python
from datetime import datetime, timedelta
from datetime import datetime as dt

def unix_to_datetime(timestamp):
    try:
        timestamp =int(timestamp)/1e9
        return pd.to_datetime(datetime.utcfromtimestamp(timestamp).strftime('%Y-%m-%d %H:%M:%S'))
    except:
        return pd.to_datetime(timestamp).strftime('%Y-%m-%d %H:%M:%S')

#unix_to_datetime('2018-05-10 06:28:53.391')
```

```python
df.loc[df.VisitDateTime.notnull(),'datetime'] = df.loc[df.VisitDateTime.notnull(),'VisitDateTime'].apply(lambda x: unix_to_datetime(x))
```

```python
df['date'] = pd.to_datetime(df.datetime).dt.date
```

### Creating a User_Vintage column to find the Vintage as of 2018-05-28

```python
df_user.loc[:,'signup_date'] = pd.to_datetime(df_user.loc[:,'Signup Date'].apply(lambda x: pd.to_datetime(x).strftime('%Y-%m-%d') ))
```

```python
df_user['today'] = pd.to_datetime( pd.to_datetime('2018-05-28').strftime('%Y-%m-%d') )
df_user['User_Vintage'] = (df_user['today'] - df_user['signup_date']).dt.days
```
